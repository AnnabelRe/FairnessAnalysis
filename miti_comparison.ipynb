This notebook uses output from fairness_results.ipynb and creates visualizations to compare mitigation strategies to baseline.

```python
import pandas as pd
import numpy as np
import os

import radar_chart
import fairness_dashboard
import fairness_dashboard_miti

import pickle 
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
import radar_chart_miti

import warnings
warnings.filterwarnings("ignore")
```

```python
output_path = "xxx" # add customized local path to output folder
miti = "post" #"in"
#groups = ["MF","MU", "FU", "AU", "ANA", "NAU"]
groups = ["AU"]

if miti == "in":
    model_names = ['Baseline LGBM', 'FTU', "FGBM", "FGBM with HPT"]
    
    for g in groups:
    
        with open(output_path+f'results_{g}.pkl', 'rb') as file:
            baseline = pickle.load(file)

        with open(output_path+f'results_nosen_{g}.pkl', 'rb') as file:
            nosen = pickle.load(file)

        with open(output_path+f'results_fgbm_{g}.pkl', 'rb') as file:
            FGBM = pickle.load(file)
        
        with open(output_path+f'results_hpt_{g}.pkl', 'rb') as file:
            hpt = pickle.load(file)
 
        radar = False
        if radar:
            baseline[1].popitem()
            nosen[1].popitem()
            FGBM[1].popitem()
            hpt[1].popitem()
            name = (output_path+f"miti_models_{g}.png")
            dict_list = [baseline[1], nosen[1], FGBM[1], hpt[1]]

            df = pd.DataFrame(dict_list, index=model_names)
            df = df.reindex(sorted(df.columns), axis=1)

            radar_chart_miti.radar_chart(df,ideal_value=1.0, filename=name)

            baseline[2].popitem()
            nosen[2].popitem()
            FGBM[2].popitem()
            hpt[2].popitem()
            name = (output_path+f"miti_models_norm_{g}.png")
            dict_list = [baseline[2], nosen[2], FGBM[2], hpt[2]]

            df = pd.DataFrame(dict_list, index=model_names)
            df = df.reindex(sorted(df.columns), axis=1)

            radar_chart_miti.radar_chart(df,ideal_value=0.5, filename=name)
            break

        bars = False
        if bars: 
            name = (output_path+f"bars_models_{g}.png")
            dict_list = [baseline[3], nosen[3], FGBM[3], hpt[3]]
            df_og = pd.DataFrame(dict_list, index=model_names)
            fairness_dashboard_miti.bar_chart(df_og, name)


            name = (output_path+f"bars_norm_models_{g}.png")
            dict_list = [baseline[4], nosen[4], FGBM[4], hpt[4]]
            df_norm = pd.DataFrame(dict_list, index=model_names)
            fairness_dashboard_miti.bar_chart(df_norm, name)
    
        
else:
    model_names = ['Baseline LGBM', 'ThresholdOpt', 'ROC', 'Eodds']#, "RejectOpt"]
    
    for g in groups:

        with open(output_path+f'results_{g}.pkl', 'rb') as file:
            baseline = pickle.load(file)

        with open(output_path+f'results_to_{g}.pkl', 'rb') as file:
            to = pickle.load(file)
        
        file_path = output_path+f'results_roc_{g}.pkl'
        if os.path.exists(file_path):
            with open(file_path, 'rb') as file:
                roc = pickle.load(file)

        file_path = output_path+f'results_eodds_{g}.pkl'
        if os.path.exists(file_path):
            with open(file_path, 'rb') as file:
                eodds = pickle.load(file)
        
        radar = False
        if radar:
            baseline[1].popitem()
            to[1].popitem()
            roc[1].popitem()
            eodds[1].popitem()

            name = (output_path+f"pp_miti_models_{g}.png")
            dict_list = [baseline[1], to[1], roc[1],eodds[1]]

            df = pd.DataFrame(dict_list, index=model_names)
            df = df.reindex(sorted(df.columns), axis=1)

            radar_chart_miti.radar_chart(df,ideal_value=1.0, filename=name)

            baseline[2].popitem()
            to[2].popitem()
            roc[2].popitem()
            eodds[2].popitem()

            name = (output_path+f"pp_miti_models_norm_{g}.png")
            dict_list = [baseline[2], to[2], roc[2],eodds[2]]

            df = pd.DataFrame(dict_list, index=model_names)
            df = df.reindex(sorted(df.columns), axis=1)

            radar_chart_miti.radar_chart(df,ideal_value=0.5, filename=name)
            
        bars = True
        if bars: 
            name = (output_path+f"pp_bars_models_{g}.png")
            dict_list = [baseline[3], to[3], roc[3],eodds[3]]
            df_og = pd.DataFrame(dict_list, index=model_names)
            fairness_dashboard_miti.bar_chart(df_og, name,miti)
            
            name = (output_path+f"pp_bars_norm_models_{g}.png")
            dict_list = [baseline[4], to[4], roc[4],eodds[4]]
            df_norm = pd.DataFrame(dict_list, index=model_names)
            fairness_dashboard_miti.bar_chart(df_norm, name,miti)
```
